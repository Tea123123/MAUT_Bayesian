```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
u_1 = [0, 0.5, 0.8, 0.9, 1]
u_2 = [0, 0.95, 0.97, 0.99, 1]
```

```python
plt.rc("font", family="serif")
plt.figure(figsize=(10, 4))

plt.subplot(121)
xticks = list(range(1, 6))    
plt.plot(x, u_1, color='black')

plt.tick_params(axis="x", direction="in")
plt.tick_params(axis="y", direction="in")

#plt.legend(loc="best")
plt.xlabel(r"$x_{j1}$")
#plt.ylabel(r"$u_1(x_{ji})$")
plt.xticks(xticks)
#plt.yticks(yticks)
plt.title(r"$u_1(x_{j1})$")
plt.grid(True)



plt.subplot(122)
xticks = list(range(1, 6))    
plt.plot(x, u_2, color='black')

plt.tick_params(axis="x", direction="in")
plt.tick_params(axis="y", direction="in")

#plt.legend(loc="best")
plt.xlabel(r"$x_{j2}$")
#plt.ylabel(r"$u_2(x_{ji})$")
plt.xticks(xticks)
#plt.yticks(yticks)
plt.title(r"$u_2(x_{j2})$")
plt.grid(True)

plt.savefig("discrete_u.eps", format="eps")
```
