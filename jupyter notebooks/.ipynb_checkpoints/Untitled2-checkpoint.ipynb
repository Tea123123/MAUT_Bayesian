```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import KFold
```

```python
from sklearn import preprocessing
```

```python
# load the data
X, y = load_breast_cancer(return_X_y=True) ## X input, y output
## to convert the {0,1} output into {-1,+1}
y = 2*y - 1
## learning parameters
nitermax = 50 ## maximum iteration
eta = 0.1 ## learniung speed
nfold = 5 ## number of folds
## to split the data into 5-folds we need
cselection = KFold(n_splits=nfold, random_state=None, shuffle=False)
```

The task is to run the Logistic Regression algorithm on the 5-fold cross-validation, and compute the F1-
score on the corresponding test sets. In selecting the training and the test sets you might follow the example
code of the KFold function provided in the Sklearn.
The question: what is the average F1 score computed on the 5-folds? Round the numbers up to 2 decimals,
and take the closest one.
Be aware, the Logistic Regression algorithm of Lecture 5 is not the same which is implemented in the
Sklearn. use that version which is presented in the Lecture!

```python
w = np.array([0]*30)

def sig(x): 
    return 1/(1 + math.exp(-x))

def j(w, x, y):
    return -sig(-y*(np.dot(w, x)))*y*x



old_y = []
pred_y = []

scores = []

for train_index, test_index in cselection.split(X2, y): 
    w = np.array([0]*30)
    
    train_x = X2[train_index]
    train_y = y[train_index]

    test_x = X2[test_index]
    test_y = y[test_index]
    
    for i in range(100000):

        ix = np.random.choice(len(train_x), 1)[0]

        n = 0.001

        dw = j(w, train_x[ix], train_y[ix])
        w = w - n*dw

        #old_y.append(train_y[ix])
        #cal = sig(np.dot(w, train_x[ix]))
        #if cal>=0:pred =1
        #else: pred = -1
        #pred_y.append(pred)

    pred = []
    for i in range(len(test_y)):   
        cal = sig(np.dot(w, test_x[i]))
        if cal >=0: pred.append(1)
        else: pred.append(-1)

    scores.append(metrics.f1_score(test_y, pred))
```
