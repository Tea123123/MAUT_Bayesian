```python
def u(x,a,s):
    if a == 0:
        return (1/6*s)*x + 0.5
    else: 
        def b(x):
            return 1-math.e**(-a*x)
        return (b(x)-b(-3*s))/(b(3*s)-b(-3*s))
def us(x_list, a, s): #utility function for the list of values           
    return [u(x, a, s) for x in x_list]
```

```python
m = 5 #number of alternatives
mu = 0
s = 1 #std for m_j
t = 1 #std for y_j
a = 0#for convex <0, concave >0


def cal_post(m, mu=0, own_s=1, btw_s=0.5, own_t=1,btw_t = 0.5, a=0):
    d_list = []
    for i in range(10000):
        #Generate $m$ true values $x_j$ from the prior
        means=np.full((1, m), mu)[0]
        x_cov = np.full((m, m), btw_s)
        np.fill_diagonal(x_cov, own_s)
        x = np.random.multivariate_normal(means, x_cov, 1)[0]
        
        #For each true value, generate an estimate $y_j$ from the likelihood
        y_cov = np.full((m, m), btw_t)
        np.fill_diagonal(y_cov, own_t)
        y = np.random.multivariate_normal(x, y_cov, 1)[0] #y_j~N(x_j,1)
        #calculate alpha
        alpha = np.matmul(x_cov, np.linalg.inv(x_cov + y_cov))
        #For each $y_j$, generate 1000 true values $x_jk$ from $(X|Y=y)=N(sigma^2/(tau^2 +sigma^2)*y_j+tau^2/(sigma^2+tau^2)*overline(mu)), k=1,...1000$
        x_p = []
        for y_j in y:

            p_means = np.add(np.matmul(alpha, y), np.matmul(np.eye(m)-alpha, means))
            p_cov = np.matmul(np.eye(m)-alpha, x_cov)
            x_p.append(np.random.multivariate_normal(p_means, p_cov, 1000).T)


        #Compute $u(x_jk$) for all $j,k$
        u_x_p = [us(x, a, own_s) for x in x_p]

        #Compute the average $\sum_{k=1}^1000 u_(x_jk)$ to get an estimate for the expected utility $E[u(X_j)|Y_j=y_j]$
        eu = [np.mean(x) for x in u_x_p]

        #Choose the alternative $j^$ with the highest expected utility
        max_eu = max(eu)
        max_j = eu.index(max_eu)

        #posterior disappointment 
        d = (u(x[max_j],a, own_s) - max_eu)
        d_list.append(d)
    #Compute the difference $u(x_\tilde{j})-\sum_{k=1}^1000 u_(x_\tilde{j}k)$
    return d_list
```

```python
d = {}
for m in [2,3,5,10]:
    d[m] = cal_post(m, a=0)
    
df = pd.DataFrame(d)
df.boxplot(showmeans=True)
plt.show()
print(df.mean(axis=0))
```

```python
def cal_post(m, mu=0, s=1, t=1, a=0):
    d_list = []
    for i in range(10000):
        #Generate $m$ true values $x_j$ from the prior
        x = np.random.normal(mu,s,m) #x_j~N(0,1)
        
        #For each true value, generate an estimate $y_j$ from the likelihood 
        y = [np.random.normal(x_j,t,1)[0] for x_j in x] #y_j~N(x_j,1)

        #For each $y_j$, generate 1000 true values $x_jk$ from $(X|Y=y)=N(sigma^2/(tau^2 +sigma^2)*y_j+tau^2/(sigma^2+tau^2)*overline(mu)), k=1,...1000$
        x_p = []
        for y_j in y:
            p_mean = (t**2/(s**2 + t**2))*mu + (s**2/(s**2 + t**2))*y_j
            p_std = (s**2*t**2/(s**2 + t**2)) #caution! it is variance

            x_p.append(list(np.random.normal(p_mean, np.sqrt(p_std),1000)))


        #Compute $u(x_jk$) for all $j,k$
        u_x_p = [us(x, a, s) for x in x_p]

        #Compute the average $\sum_{k=1}^1000 u_(x_jk)$ to get an estimate for the expected utility $E[u(X_j)|Y_j=y_j]$
        eu = [np.mean(x) for x in u_x_p]

        #Choose the alternative $j^$ with the highest expected utility
        max_eu = max(eu)
        max_j = eu.index(max_eu)

        #posterior disappointment 
        d = (u(x[max_j],a, s) - max_eu)
        d_list.append(d)
    #Compute the difference $u(x_\tilde{j})-\sum_{k=1}^1000 u_(x_\tilde{j}k)$
    return d_list
```

```python
def cal_post_ver2(m, mu=0, s=1, t=1, a=0):
    d_list = []
    for i in range(10000):
        #Generate $m$ true values $x_j$ from the prior
        x = np.random.normal(mu,s,m) #x_j~N(0,1)
        
        #For each true value, generate an estimate $y_j$ from the likelihood 
        y = [np.random.normal(x_j,t,1)[0] for x_j in x] #y_j~N(x_j,1)

        #For each $y_j$, generate 1000 true values $x_jk$ from $(X|Y=y)=N(sigma^2/(tau^2 +sigma^2)*y_j+tau^2/(sigma^2+tau^2)*overline(mu)), k=1,...1000$
        x_p = []
        for y_j in y:
            p_mean = (t**2/(s**2 + t**2))*mu + (s**2/(s**2 + t**2))*y_j
            p_std = (s**2*t**2/(s**2 + t**2)) #caution! it is variance

            x_p.append(list(np.random.normal(p_mean, np.sqrt(p_std),1000)))


        #Compute $u(x_jk$) for all $j,k$
        u_x_p = [us(x, a, s) for x in x_p]

        #Compute the average $\sum_{k=1}^1000 u_(x_jk)$ to get an estimate for the expected utility $E[u(X_j)|Y_j=y_j]$
        eu = [np.mean(x) for x in u_x_p]

        #Choose the alternative $j^$ with the highest expected utility
        max_eu = max(eu)
        max_j = eu.index(max_eu)

        #posterior disappointment 
        d = (u(max(x),a, s) - max_eu)/max_eu
        d_list.append(d)
    #Compute the difference $u(x_\tilde{j})-\sum_{k=1}^1000 u_(x_\tilde{j}k)$
    return d_list
```
